```python
data_p = data.copy()
data_p.drop(columns=['f16'], inplace=True)
data_p.drop(columns=['f15'], inplace=True)
data_p.drop(columns=['f14'], inplace=True)
data_p.drop(columns=['f13'], inplace=True)
data_p.drop(columns=['f12'], inplace=True)
data_p.drop(columns=['f10'], inplace=True)
data_p.replace('no', 0, inplace=True)
data_p.replace('yes', 1, inplace=True)
data_p.drop(columns=['p'], inplace=True)
```

```python
# plot correlation
corr = data_p.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import tensorflow as tf
```

```python
tf.config.list_physical_devices('GPU')
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.metrics import Precision, Recall
```

```python
model = Sequential()
model.add(Dense(16, input_dim=11, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
adam = Adam(learning_rate=0.001)
model.compile(loss='binary_crossentropy', optimizer=adam,  metrics=[
        Precision(),
        Recall()
    ])
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
train_val_df, test_df = train_test_split(data_p, test_size=0.2, random_state=42)

train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=42)  

```

```python
train_y = train_df['target']
train_x = train_df.drop(columns=['target'])

val_y = val_df['target']
val_x = val_df.drop(columns=['target'])

test_y = test_df['target']
test_x = test_df.drop(columns=['target'])


scaler = MinMaxScaler()
train_x = scaler.fit_transform(train_x)
val_x = scaler.transform(val_x)
test_x = scaler.transform(test_x)
```

```python
model.fit( train_x, train_y, validation_data=(val_x, val_y), epochs=5)
```
